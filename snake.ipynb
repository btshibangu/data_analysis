```python
import pygame
from pygame.locals import *
import random

pygame.init()

largeur_fenetre = 800
hauteur_fenetre = 600

fenetre = pygame.display.set_mode((largeur_fenetre, hauteur_fenetre))
pygame.display.set_caption('Snake Game')
fenetre.fill((255, 255, 255))
pygame.display.update()

def jeu_snake():
    # Initialisation du serpent
    serpent = [(largeur_fenetre // 2, hauteur_fenetre // 2)]
    direction = (0, 1)  # Départ vers le bas
    score = 0

    # Génération aléatoire de la pomme
    pomme = (random.randint(0, largeur_fenetre-10), random.randint(0, hauteur_fenetre-10))

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    running = False
                elif event.key == pygame.K_UP and direction != (0, 1):  # Ne pas permettre le mouvement inverse
                    direction = (0, -1)
                elif event.key == pygame.K_DOWN and direction != (0, -1):
                    direction = (0, 1)
                elif event.key == pygame.K_LEFT and direction != (1, 0):
                    direction = (-1, 0)
                elif event.key == pygame.K_RIGHT and direction != (-1, 0):
                    direction = (1, 0)

        # Gestion des mouvements du serpent
        tête = serpent[0]
        nouvelle_tête = (tête[0] + direction[0] * 10, tête[1] + direction[1] * 10)
        serpent.insert(0, nouvelle_tête)

        # Détection des collisions avec les bords de la fenêtre
        if nouvelle_tête[0] < 0 or nouvelle_tête[0] >= largeur_fenetre or nouvelle_tête[1] < 0 or nouvelle_tête[1] >= hauteur_fenetre:
            running = False

        # Détection de la collision avec la pomme
        if nouvelle_tête == pomme:
            score += 1
            # Générer une nouvelle position de pomme
            pomme = (random.randint(0, largeur_fenetre-10), random.randint(0, hauteur_fenetre-10))

        # Mise à jour de l'affichage
        fenetre.fill((255, 255, 255))

        # Dessiner le serpent
        for partie in serpent:
            pygame.draw.rect(fenetre, (0, 255, 0), (partie[0], partie[1], 10, 10))

        # Dessiner la pomme
        pygame.draw.rect(fenetre, (255, 0, 0), (pomme[0], pomme[1], 10, 10))

        # Afficher le score
        font = pygame.font.Font(None, 36)
        score_text = font.render("Score: " + str(score), True, (0, 0, 0))
        fenetre.blit(score_text, (10, 10))

        pygame.display.update()

        # Gérer la vitesse du jeu
        pygame.time.delay(100)  # 100 millisecondes

    pygame.quit()

jeu_snake()
```
