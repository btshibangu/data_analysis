```python
import cv2
import numpy as np
import csv

# Load the image
image = cv2.imread("./image.jpeg")
# Extract pixel values
pixel_values = image.reshape(-1, 3)  # Assumes the image is in RGB color space

# Write to CSV file
with open('image.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Red', 'Green', 'Blue'])
    writer.writerows(pixel_values)
```
